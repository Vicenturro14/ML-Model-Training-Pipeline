# Notebook to sample de dataset

```python
# Lets create a dataset with the file
import pandas as pd

# Create a dictionary with data
arxiv_data = pd.read_json('data/arxiv.json', lines=True)
```

```python
# We take a sample of 1000 articles
arxiv_finetune_sample = arxiv_data.sample(200)
arxiv_test_sample = arxiv_data.sample(1000)

# Save the sample to a csv file
arxiv_finetune_sample.to_csv('data/arxiv_finetune_sample.csv', index=False)
arxiv_test_sample.to_csv('data/arxiv_test_sample.csv', index=False)
```
